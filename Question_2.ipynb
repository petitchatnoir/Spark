#9. Suivre l'√©volution des ventes et identifier les tendances de vente au fil du temps

```
%sql
SELECT 
  year, 
  month, 
  SUM(sales) as total_sales, 
  CONCAT(ROUND(100 * (SUM(sales) / LAG(SUM(sales), 1, 0) OVER (ORDER BY year, month) - 1), 2), "%") as sales_growth 
FROM global_temp.produits 
GROUP BY year, month 
ORDER BY year, month
```
